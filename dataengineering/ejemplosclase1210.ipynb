```python
numero_decimal = 132  # Reemplaza 42 con el número decimal que deseas convertir
numero_binario = bin(numero_decimal)

# La variable numero_binario contendrá una cadena que representa el número en binario, con el prefijo '0b'
# Para mostrar solo el valor binario sin el prefijo, puedes hacer lo siguiente:
valor_binario = numero_binario[2:]

print(f'Número decimal: {numero_decimal}')
print(f'Número binario: {valor_binario}')
```
