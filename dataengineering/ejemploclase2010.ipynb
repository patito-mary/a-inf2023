```python
a = np.random.normal(size=150)
```

```python
df = pd.DataFrame(a)
df.to_csv('~/gitcosas/a-inf2023/dataengineering/df.txt')
```

Para crear o leer un archivo plano en python, se usa la funcion buildint `open`

Esta funcion nos permite obtener $\textbf{objetos tipo archivo!}$

debido a que siempre hay que cerrar los archivos, lo ideal es ocupar la funcion `with`

```python
with open('datos.txt', 'w') as f:
    for i in a:
        f.write('{:.6f}\n'.format(i))
```

```python
with open('datosconfstring.txt', 'w') as f:
    for i in range(len(a)):
        f.write(f'{i}\t{a[i]:.5f}\n')
```

Metodo para leer archivos de texto de lineas arbitrarias


$\textbf{Recordar siempre abrir los archivos!}$

```python
f = open('datosconfstring.txt', 'r')
ar = f.readlines()
```

Como transformo de linea de string a numero?
`split` lee caracter por caracter y junto con `float` sirven para tranformar las lineas de string a flotante

```python
b[0] = float(ar[0].split()[0])
b[1] = float(ar[0].split()[1])
```
