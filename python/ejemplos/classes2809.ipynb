```python
class Galleta:
    def __init__(self, sabor='chocolate', tipo='rellena'):
        self.sabor = sabor
        self.tipo = tipo
    def mostrar(self):
        print(f'Mi galleta es de {self.sabor} y viene {self.tipo}')
```

```python
galletita = Galleta('frutilla', 'rellena')
galletita.mostrar()
Galleta.mostrar(galletita)
```

```python
type(galletita.mostrar())
type(Galleta.mostrar(galletita))
```
