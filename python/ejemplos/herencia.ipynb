```python
class Animal:
    """Clase base para crear mas clases vivas"""
    def __init__(self, nombre):
        self.nombre = nombre
    def hablar(self):
        pass
```

```python
class Gato(Animal):
    def hablar(self):
        return 'miau miau'

class Perro(Animal):
    def hablar(self):
        return 'guau'
    
```
