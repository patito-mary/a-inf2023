```python
class Complex:
    """Clase de numeros complejos mejorada"""
    def __init__(self,real,imaginary):
        """Constructor"""
        self.r = real
        self.i = imaginary

    def __str__(self):
        return str(self.r)+'+i'+str(self.i)

    def __add__(self, c):
        print(self)
        return Complex(self.r + c.r, self.i+c.i)
```
